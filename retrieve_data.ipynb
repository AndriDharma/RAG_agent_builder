```python
column_data = df['text']
num_rows = len(column_data)
rows_per_file = 1000
output_dir = r"C:\Users\ASUS\Documents\Job\Datalabs\2025\specialization\data\text"
num_files = (num_rows + rows_per_file - 1) // rows_per_file  # Calculate number of files

for file_idx in range(num_files):
    start_row = file_idx * rows_per_file
    end_row = min((file_idx + 1) * rows_per_file, num_rows)

    filename = os.path.join(output_dir, f"output_{file_idx + 1}.txt")
    with open(filename, 'w') as f:
        for row_idx in range(start_row, end_row):
            f.write(str(column_data.iloc[row_idx]) + '\n')
```
